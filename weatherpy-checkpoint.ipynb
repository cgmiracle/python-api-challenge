## Generate Cities List

```python
#Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import time
import requests
from datetime import date
import numpy as np
from pprint import pprint
from api_keys import weather_api_key
from citipy import citipy
from scipy.stats import linregress 
import os

#Create a location for output file
output_file = "output/cities.csv"

#Set a range of Lat and Long
lat = (-90, 90)
long = (-180, 180)
```

```python
os.chdir(r'C:\Users\Cody\UCBE\python-api-challenge')
```

```python
#Associate lat and long with nearest cities.
#Create lists for holding lat and long
lat_long = []
cities = []

#Randomize cities list
lats = np.random.uniform(lat[0], lat[1], size=1500)
longs = np.random.uniform(long[0], long[1], size=1500)
lat_long = zip(lats, longs)

#Identify nearest city
for lat_longs in lat_long:
    city = citipy.nearest_city(lat_longs[0], lat_longs[1]).city_name
    
    #Check if city is unique
    if city not in cities:
        cities.append(city)
        
#Check city count
print(len(cities))
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
#Create URL for API call
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=cd29c6e3a218049e216c1b1fb8be334a"

#City data
citydata = []

#Print city data to log
print("City Data Retrieval Initiated   ")
print("--------------------------------")

#counters
record_count = 1
set_count = 1

#Create loop for all cities in list
for i, city in enumerate(cities):
    if (i % 50 == 0 and i >= 50):
        set_count += 1
        record_count = 0
        
    #Create endpoint URL for cities
    city_url = url + '&q=' + city
    
    print("Processing Record %s of Sets %s | %s" % (record_count, set_count, city))
    
    #add to record count
    record_count += 1
    
    try:
        city_weather = requests.get(city_url).json()
        
        #Parse out relevent data
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        city_country = city_weather["sys"]["country"]
        city_date = city_weather["dt"]
        
        citydata.append({"City": city, 
                          "Lat": city_lat, 
                          "Long": city_lng, 
                          "Max Tempurature": city_max_temp,
                          "Humidity": city_humidity,
                          "Cloudiness": city_clouds,
                          "Wind Speed": city_wind,
                          "Country": city_country,
                          "Date": city_date})
    
    #Exception for if there is an error, skip city
    except:
        print("City not found.")
        pass
        
print("------------------")
print("City Data Compled ")
print("------------------")
```

## Linear Regression

```python
#Linear regression function
def plot_linear_regression(x_values, y_values, title, text_coordinates):
    #S. Hem regression
    (slope, intercept, rvalue, pvalie, stderr) = linregress(x_values, y_values)
    regress_values = x_values * slope + intercept
    line_eq = 'y = ' + str(round(slope,2)) + 'x + ' + str(round(intercept,2))
    
    #Plot regression
    plt.scatter(x_values,y_values)
    plt.plot(x_values,regress_values,"r-")
    plt.annotate(line_eq,text_coordinates,fontsize=15,color="red")
    plt.xlabel("Latitude")
    plt.ylabel(title)
    print(f"The r-value is: {rvalue**2}")
    plt.show()
```

```python
#DataFrames for N. Hem and S. Hem
n_hem = citydata_df.loc[(citydata_df["Lat"] >=0)]
s_hem = citydata_df.loc[(citydata_df["Lat"] <0)]
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
x_values = n_hem["Lat"]
y_values = n_hem["Max Tempurature"]
plot_linear_regression(x_values, y_values, "Max Tempurature", (-55, 90))
```

####  Southern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
x_values = s_hem["Lat"]
y_values = s_hem["Max Tempurature"]
plot_linear_regression(x_values, y_values, "Max Tempurature", (-55, 90))
```
